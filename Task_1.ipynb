## Step - I
### <span style="color:red">IMPORTING ALL THE REQUIRED PACKAGES</span>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

## Step - II
### <span style="color:red">IMPORTING THE GIVEN DATA SET</span>

```python
Url = "http://bit.ly/w-data"
df = pd.read_csv(Url)
print("Data imported")
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

## Step - III
### <span style='color:red'>VISUALIZE AND ANALYSE THE DATASET</span>

```python
df.plot(x='Hours', y='Scores', style='ro', markersize='7')
plt.title("Hours Vs Percentage")
plt.xlabel("Hours Studied")
plt.ylabel('Percentage Scored')
plt.show()
```

```python
df.corr()
```

## Step - IV
### <span style='color:red'>PREPARING OUR DATA</span>

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
```

```python
x
```

```python
y
```

## Step - V
### <span style="color:red">SPLITTING THE DATA FOR TESTING</SPAN>

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

## STEP - VI
### <span style='color:red'>PLOTTING REGRESSION LINE</span>

```python
regresor = LinearRegression()
regresor.fit(x_train, y_train)
```

```python
line = regresor.coef_*x+regresor.intercept_
plt.scatter(x, y)
plt.plot(x, line)
plt.show()
```

## Step - VII
### <span style='color:red'>MAKING THE PREDICTIONS</span>

```python
y_predict = regresor.predict(x_test)
df = pd.DataFrame({'Actual': y_test, 'predicted': y_predict})
df
```

## Step - VIII
### <span style='color:red'>VISUALIZE THE COMPARISION BETWEEN ACTUAL AND PREDICTED SCORES</span>

```python
df.plot(kind='bar',figsize=(7,7))
plt.show
```

## STEP-6 
### <span style='color:red'>TO MAKE PREDICTIONS<span>

```python
hours = 9.25
own_predict = regresor.predict([[hours]])
print("No. of hours studied by student", hours)
print("Predicted percentage of the student =", own_predict[0])
```

## Step - IX
### <span style='color:red'>MODEL EVALUATION<span>

```python

print("Mean Absolute Error :-", metrics.mean_absolute_error(y_test,y_predict))
print("Mean Squared Error :-", metrics.mean_squared_error(y_test,y_predict))
print("Root Mean Squared Error :-", np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
```
